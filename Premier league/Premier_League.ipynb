```python
import pandas as pd

def comparar_columnas(df1, df2):
    # Obtener las columnas de cada DataFrame
    columnas_df1 = set(df1.columns)
    columnas_df2 = set(df2.columns)

    # Columnas comunes
    comunes = columnas_df1.intersection(columnas_df2)

    # Columnas nicas a cada DataFrame
    unicas_df1 = columnas_df1 - comunes
    unicas_df2 = columnas_df2 - comunes

    print("Columnas en ambos DataFrames:")
    print(sorted(comunes))
    print("\nColumnas nicas en el primer DataFrame:")
    print(sorted(unicas_df1))
    print("\nColumnas nicas en el segundo DataFrame:")
    print(sorted(unicas_df2))

# Ejemplo de uso con tus DataFrames df5 y df6
comparar_columnas(df5, df6)
```
