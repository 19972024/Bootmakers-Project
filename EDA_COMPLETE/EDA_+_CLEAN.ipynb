```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar solo las columnas numricas para el anlisis de outliers
numeric_columns = clean_data.select_dtypes(include=['number']).columns

# Crear subconjunto de datos solo con columnas numricas
data_numeric = clean_data[numeric_columns]

# Configurar el tamao de la figura
plt.figure(figsize=(15, 20))

# Iterar sobre cada columna y crear un boxplot
for i, col in enumerate(data_numeric.columns):
    plt.subplot(8, 4, i + 1)  # Ajustar el nmero de filas y columnas segn la cantidad de columnas seleccionadas
    sns.boxplot(x=data_numeric[col])
    plt.title(col)
    plt.tight_layout()

# Mostrar los grficos
plt.show()
```

```python
#Identificar Outliers: Los boxplots son especialmente tiles para detectar outliers. Los puntos fuera de los "bigotes" del boxplot son considerados outliers.
#Observa la forma de la caja y la longitud de los bigotes para entender la distribucin de los datos. Una caja larga indica una alta variabilidad en los datos, mientras que una caja corta indica baja variabilidad.
#Columnas con muchos outliers: FTAG, FTHG, AvgCD, etc. Esto podra indicar la presencia de valores atpicos o errores de entrada de datos o si representan variaciones naturales.
#Alta variabilidad: Columnas como FTAG y FTHG tienen cajas largas, indicando que hay una alta variabilidad en los goles anotados en los partidos.
```

```python
#Decidir no modificar, eliminar ni aplicar transformaciones a las variables numricas es una decisin informada basada en el anlisis de las distribuciones de los datos. 
#Esta decisin se fundamenta en la necesidad de mantener la integridad y representatividad de los datos originales, respetando su variabilidad natural y asegurando que los modelos construidos reflejen fielmente la realidad observada. 
#De este modo, se evita la introduccin de sesgos y se preserva toda la informacin valiosa que los datos pueden ofrecer para anlisis y predicciones futuras.
```
